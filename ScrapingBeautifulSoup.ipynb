```python

import requests
from bs4 import BeautifulSoup
import pandas as pd
pd.options.display.max_columns=None
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
comp=[]
rol=[]
loc=[]
dat=[]
emp_type=[]
job_desc=[]
job_link=[]
no_appe=[]
for i in ['Data scientist','Data Engineer','Data Analyst','Data','Machine Learning Engineer']:
    keyword=i
    print(f"Scraping Role {keyword}")
    base_url = "https://www.linkedin.com/jobs/search/?currentJobId=3597020723&f_E=1%2C2%2C3&geoId=102713980&keywords="+keyword+"%20&location=India&refresh=true"
    pages = 10
    page_size = 10

    def no_of_applicants(job_link):
        a2=None
        response1 = requests.get(job_link)
        content1 = response1.content
        parsed_content1 = BeautifulSoup(content1, 'html.parser')
        a2 = parsed_content1.find('span',{'class':'num-applicants__caption topcard__flavor--metadata topcard__flavor--bullet'})
        return (a2)

    def employement_type(job_link):
        a1=None
        response1 = requests.get(job_link)
        content1 = response1.content
        parsed_content1 = BeautifulSoup(content1, 'html.parser')
        job_description = parsed_content1.find_all('ul',{'class':'description__job-criteria-list'})   
        for i in job_description:
            a1=(i.find('span',{'class':'description__job-criteria-text description__job-criteria-text--criteria'}).text.strip())
        return a1  

    for i in range(1, pages + 1):
        print(f"Scraping page {i}")
        url = f"{base_url}/page/{i}/?sortby=post_date%3ADesc&pagesize={page_size}"
        response = requests.get(url)
        content = response.content
        parsed_content = BeautifulSoup(content, 'html.parser')
        job_list = parsed_content.find('ul', class_='jobs-search__results-list')
        if job_list is not None:
            job_items = job_list.find_all('li')

        for i in job_items:
            if i.find('h4',{'class':"base-search-card__subtitle"}):
                comp.append(i.find('h4',{'class':"base-search-card__subtitle"}).text.strip())
            else:
                comp.append('NA')
            if i.find('h3'):
                rol.append(i.find('h3').text.strip())
            else:
                rol.append('NA')
            if i.find('span',{'class':'job-search-card__location'}):
                loc.append(i.find('span',{'class':'job-search-card__location'}).text.strip())
            else:
                loc.append('NA')
            if i.find('time',{'class':"job-search-card__listdate"}):
                dat.append(i.find('time',{'class':"job-search-card__listdate"}).text.strip())
            else:
                dat.append('NA')
            if i.find('a', {'class': 'base-card__full-link'}):
                
                job_lin = i.find('a', {'class': 'base-card__full-link'})['href']
                
                job_link.append(job_lin)
            else:
                job_link.append(None)
            
            if job_link[-1]!=None:
                emp_typ = employement_type(job_link[-1])
                no_app=no_of_applicants(job_link[-1])
                no_appe.append(no_app)
            else:
                emp_typ = None
                no_appe.append(None)
            if emp_typ is not None:
                emp_type.append(emp_typ)
            else:
                emp_type.append("NA")
            
            
        print(f"   ---> {len(emp_type)} total jobs")
```

```python
dict={'Company':comp,
      'Role':rol,
     'Location':loc,
     'Date_Posted':dat,
     'Job_link':job_link,
     'Employment_Type':emp_type,
     'No_of_applicants':no_appe}    
df=pd.DataFrame(dict)
```
